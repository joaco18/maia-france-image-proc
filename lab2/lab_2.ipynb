# 2.1 Quantization

## 2.1.1.1
Matlab'd quant() function quant(X,Q) takes two inputs and returns values from X rounded to nearest multiple of Q.

```python
def quant(value: np.ndarray, quantity: int):
    """ 
    Discretize values to the closest multiple of quantity.
    Args:
        value (np.ndarray or float): value to quantize
        value (np.ndarray or float): integer to get the closest values from.
    Returns:
        discretized value or the nd.array with the discretization applied elementwise.
    """
    res = (value - (value % quantity)) + np.round((value % quantity) / quantity) * quantity
    return res.astype(int)
```

## 2.1.1.2

```python
img = cv2.imread("../lab1/images/saturn.png", cv2.IMREAD_GRAYSCALE)
```

```python
def quantize(im: np.ndarray, n: int=None, b: int = None):
    """
    Args:
        im (np.ndarray): gray-scale image to quatize.
        n (int): number of gray levels.
        b (int): number of bits. If b is passed, n is not used.
    Returns:
        (np.ndarray) quantized image.
    """
    if (n is None) and (b is None):
        print('One of n or b shoul be passed')
        return None
    
    int_range = np.max(im) - np.min(im)
    
    if b is not None:
        int_step = np.round(int_range/((2**b)-1))
    else:
        int_step = np.round(int_range/n)
        
    res = quant(im, int_step)
    return res
```

```python
fig, ax = plt.subplots(1,2,figsize = (10,15))
ax[0].set_title('Original')
ax[0].imshow(img, cmap='gray')
ax[0].axis('off')
ax[1].set_title('Quantized')
ax[1].imshow(quantize(np.array(img), b=1), cmap='gray')
ax[1].axis('off')
plt.show()
```

```python
im = cv2.imread('../lab2/images/barbara.png', cv2.IMREAD_GRAYSCALE)
k = 1/25 * np.ones((5,5))
blured = cv2.filter2D(src=im, ddepth=-1, kernel=k)

fig, ax = plt.subplots(2, 5, figsize=(15,7))
for i in range(2,11,2):    
    title = f'Subsample x{i} \nover original' if i != 2 else 'Original'
    k2 = int(i/2)-1
    ax[0, k2].set_title(title)
    ax[0, k2].imshow(im[::i, ::i], cmap='gray')
    ax[0, k2].axis('off')
    
    title = f'Subsample x{i} \nover blured MSE' if i != 1 else 'Blured'
    ax[1, k2].set_title(title)
    ax[1, k2].imshow(blured[::i, ::i], cmap='gray')
    ax[1, k2].axis('off')
plt.show()
```

It can be seen that bluring the image before downsampling allows us to retain more information in the downsampled image. This occures since the information of each pixel in the original image is first is distributed across the sorrounding pixels, so the downsampling doesn't fully remove the information from those pixels.

In particular in regions of great details with fast changes, as in the pants, the downsampling results in the generation of fake paterns that are avoided if first we apply the average filter.

Another way to compare the information contained in each of the subsampled images is to apply the same resizing (interpolation) method to both of them and computing for example de MSE with the original image. If more information is preserved doing the mean filtering the MSE should be less.

```python
im = cv2.imread('../lab2/images/barbara.png', cv2.IMREAD_GRAYSCALE)
k = 1/25 * np.ones((5,5))

blured = cv2.filter2D(src=im, ddepth=-1, kernel=k)

results = []
for i in range(2,11,2):    
    down_res = cv2.resize(im[::i, ::i], (im.shape[1], im.shape[0]))
    blured_res = cv2.resize(blured[::i, ::i], (im.shape[1], im.shape[0]))
    
    mse_down_res = np.mean((im - down_res)**2)
    mse_blured_res = np.mean((im - blured_res)**2)
    
    results.append([mse_down_res, mse_blured_res])
pd.DataFrame(results, index=np.arange(2,11,2), columns=['MSE w/o blurring', 'MSE w blurring'])
```

By doing this, we can see that, even if visually we can see an improventen in the resizing. A very similar (or even) larger information loss occures in the blurred case.
